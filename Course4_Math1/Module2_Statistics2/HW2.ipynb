```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Домашнее_задание_2.xlsx', sheet_name='задание 2', header=1)
df = df.iloc[:7,:]
df
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(df['share, %'][:-1])
plt.show()
```

### На визуализации в примере показаны доли компаний не включая доли "других" компаний, что реально не отражает доли компаний на всем рынке

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(df['share, %'], labels=(df['share, %'] * 100).round(1).astype(str) + '%')
plt.show()
```

### В действительности большая часть рынка принадлежит "другим компаниям", а компании показанные на первом графике действительно близки по доле в рынке, но теперь лучше по моему мнению видно разницу и они в действительности составляют намного меньшую доли на всем рынке

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.barplot(data=df.sort_values('revenue, $M', ascending=False), x='company', y='revenue, $M', ax=ax)
plt.show()
```
